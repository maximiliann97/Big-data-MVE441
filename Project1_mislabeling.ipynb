### Split and mislabel the dataset

```python
from sklearn.model_selection import train_test_split

#label are the value we want to predict
labels=raw_data['RainTomorrow']
pca = PCA(n_components=5)
levels = np.linspace(0, 0.7, num=8)
iterations = 10

def split_mislabel(data, labels, level):
    data = data.drop('RainTomorrow',axis=1)
    
    # Split the data into training and testing sets
    # Split to get unaltered test set
    train_features, test_features, train_labels, test_labels = train_test_split(data, labels, stratify=labels, test_size=0.20)
    
    if level == 0:
        train_features_pca = pca.fit_transform(train_features)
        test_features_pca = pca.fit_transform(test_features)
    
    else:
        # Calculate the number of labels to swap based on the percentage
        n_swap = int(len(train_labels) * level)

        # Randomly select the indices of the labels to be swapped
        swap_indices = np.random.choice(len(train_labels), n_swap, replace=False)
        train_labels.iloc[swap_indices] = 1 - train_labels.iloc[swap_indices]


        train_features_pca = pca.fit_transform(train_features)
        test_features_pca = pca.fit_transform(test_features)


    return train_features_pca, test_features_pca, train_labels, test_labels
```

### Plot function

```python
import seaborn as sns
def plot_sns(score_matrix, levels, score, method):
    df = pd.DataFrame(score_matrix.T)
    percentages = [str(np.ceil(i * 100)) + '%' for i in levels]
    df.columns = percentages
    f = sns.pointplot(data=df)
    f.set(xlabel='Mislabeling level', ylabel=score, title=method)
```

## KNN

```python
from sklearn import neighbors
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score, accuracy_score, roc_auc_score
```

```python
best_k = 31 #Best k found to be 31 through cross-validation
knn_f1_scores = np.zeros((len(levels), iterations))
knn_accuracy_scores = np.zeros((len(levels), iterations))
knn_roc_auc_scores = np.zeros((len(levels), iterations))

for i, level in enumerate(levels):
    for j in range(iterations):
        train_features_pca, test_features_pca, train_labels, test_labels = split_mislabel(raw_data, labels, level)

        # Train a K-nearest neighbor classifier on the projected data for the three different levels
        knn = KNeighborsClassifier(n_neighbors=best_k).fit(train_features_pca, train_labels)
        
        # predict the class labels for the test set
        y_pred = knn.predict(test_features_pca)
        
        # compute the evaluation metrics and append to the respective list
        knn_f1_scores[i, j] = f1_score(test_labels, y_pred)
        knn_accuracy_scores[i, j] = accuracy_score(test_labels, y_pred)
        knn_roc_auc_scores[i, j] = roc_auc_score(test_labels, y_pred)
    

knn_f1 = np.mean(knn_f1_scores, axis=1)
knn_accuracy = np.mean(knn_accuracy_scores, axis=1)
knn_roc_auc = np.mean(knn_roc_auc_scores, axis=1)
```

```python
score_list =  [knn_f1_scores, knn_accuracy_scores, knn_roc_auc_scores]
score_names = ['F1', 'Accuracy', 'ROC AUC']
method = 'KNN'

for score_name, score in zip(score_names, score_list):
    plot_sns(score, levels, score_name, method)
    plt.show()
```

```python
# Creating mislabeling function for mislabeling one class more
def split_mislabel_2(data, labels, level, level2):
    data = data.drop('RainTomorrow',axis=1)

    # Split the data into training and testing sets
    # Split to get unaltered test set
    train_features, test_features, train_labels, test_labels = train_test_split(data, labels, stratify=labels, test_size=0.20)

    if level == 0:
        train_features_pca = pca.fit_transform(train_features)
        test_features_pca = pca.fit_transform(test_features)

    else:
        # Calculate the number of labels to swap based on the percentage
        train_labels = train_labels.to_numpy()
        n_swap = int(len(train_labels) * level)
        num_zeros = np.count_nonzero(train_labels==0)
        num_ones = np.count_nonzero(train_labels==1)

        x = (n_swap - level2 * num_zeros)/(num_ones*level)
        misclassified_zeros = int(level2 * num_zeros)
        misclassified_ones = int(level * x * num_ones)

        
        
        indices1 = np.where(train_labels == 0)[0]  # Get the indices of all the zeros
        misclassified_indices1 = np.random.choice(indices1, size=misclassified_zeros, replace=False)  # Select a random sample of indices to misclassify
        train_labels[misclassified_indices1] = 1
        
        indices2 = np.where(train_labels == 1)[0]  # Get the indices of all the zeros
        misclassified_indices2 = np.random.choice(indices2, size=misclassified_ones, replace=False)  # Select a random sample of indices to misclassify
        train_labels[misclassified_indices2] = 0


        train_features_pca = pca.fit_transform(train_features)
        test_features_pca = pca.fit_transform(test_features)


    return train_features_pca, test_features_pca, train_labels, test_labels
```

```python
level = 0.4
level2 = 0.2
knn2_f1 = []
knn2_acc = []
knn2_roc = []

for i in range(iterations):
    train_features_pca, test_features_pca, train_labels, test_labels = split_mislabel_2(raw_data, labels, level, level2)
        
    # Train a K-nearest neighbor classifier on the projected data for the three different levels
    knn = KNeighborsClassifier(n_neighbors=best_k).fit(train_features_pca, train_labels)

    # predict the class labels for the test set
    y_pred = knn.predict(test_features_pca)

    # compute the evaluation metrics and append to the respective list
    knn2_f1.append(f1_score(test_labels, y_pred))
    knn2_acc.append(accuracy_score(test_labels, y_pred))
    knn2_roc.append(roc_auc_score(test_labels, y_pred))
    

knn2_f1_mean = np.mean(knn2_f1)
knn2_acc_mean = np.mean(knn2_acc)
knn2_roc_mean = np.mean(knn2_roc)
```
